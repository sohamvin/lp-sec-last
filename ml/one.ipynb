```python
import math

# add distance between pickup location and dropoff location
# in the df

def euclidean_distance(lat1, lon1, lat2, lon2):
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    distance = math.sqrt(dlat**2 + dlon**2)
    return distance
```

```python
df['distance'] = df.apply(
    lambda row : euclidean_distance(
        row['pickup_latitude'], 
        row['pickup_longitude'], 
        row['dropoff_latitude'], 
        row['dropoff_longitude']
    ),
    axis = 1
)


df.head()
```

```python
def remove_outliers(df, feat):
    q1, q3 = np.percentile(df[feat], [25, 75])
    iqr = q3 - q1
    df_cleaned = df[(df[feat] >= q1 - 1.5*iqr) & (df[feat] <= q3 + 1.5*iqr)]
    return df_cleaned
```
