```python
def func(x):
    return (x+3)**2

def derivative(x):
    return 2*(x+3)**1

```

```python
grads = []

def grad_descent(starting_x, learning_rate, num_its):

    x = starting_x
    
    for i in range(num_its):
        
        x = x - learning_rate* derivative(x)
        
        grads.append(x)

        print(f"Iteration {i+1}: X : {x} , F(X) : {func(x)} , Derivative of X : {derivative(x)}")

    return x


        
```

```python
starting_x = 5
learning_rate = 0.1
num_iterations = 50
```

```python
min_x = grad_descent(starting_x, learning_rate, num_iterations)
print(f"Local minimum occurs at x = {min_x}")
```
