```python
import pandas as pd
```

```python
df = pd.read_csv('2-emails.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop('Email No.', inplace=True, axis=1)
```

```python
from sklearn.model_selection import train_test_split

Y = df['Prediction']
X = df.drop(['Prediction'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=500)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier()

classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
help(confusion_matrix)
```

```python
confusion_matrix(Y_test, y_pred)
```

```python
print(classification_report(Y_test, y_pred))
```

What is an SVM?
A Support Vector Machine is a classifier that finds the optimal separating hyperplane between classes. Let me break down what that means:
Hyperplane: In 2D, this is a line. In 3D, it's a plane. In higher dimensions (like your word vector space), it's called a hyperplane. It's just the decision boundary that separates your classes.
Optimal: This is the key insight. There are infinitely many lines you could draw to separate two classes, but SVM finds the one that maximizes the margin (the distance) between the classes.

The Math: Linear SVM
For a binary classification problem, we want to find a hyperplane defined by:
w · x + b = 0
where:

w is the weight vector (perpendicular to the hyperplane)
x is your input vector (your email's word vector)
b is the bias term (shifts the hyperplane away from origin)
· represents the dot product

For a point to be classified as class +1 (say, spam), we want: w · x + b ≥ 1
For class -1 (not spam), we want: w · x + b ≤ -1
The margin (the distance between the hyperplane and the nearest points) turns out to be 2/||w||, where ||w|| is the magnitude of the weight vector.
To maximize the margin, we need to minimize ||w||, or more commonly, minimize (1/2)||w||² (the squared version is used because it's mathematically nicer to work with and gives the same solution).

kernel='linear': Uses a linear decision boundary. The classifier finds a straight hyperplane (w · x + b = 0) to separate your email classes. This is the simplest kernel - no fancy transformations, just finds the best linear separator in your word vector space.

```python
svm = SVC(kernel='linear', probability=True) 
```

```python
svm.fit(X_train, Y_train)
y_pred_svm = svm.predict(X_test)
y_pred_svm
```
